## **02 Random Walk**
---

Here we implement the random walk benchmark, which is used for comparison against all other models.

This forecast sets $\hat{y}_{t+h} = y_t$:  
the prediction of a future value is simply the current value.

This is the standard benchmark for yield curve forecasting and is famously
difficult to beat (Diebold & Li, 2006).

We evaluate RW forecasts on the three Nelson-Siegel factors at horizons h = 1, 3, 6, and 12 months ahead [kind of arbitrary for now - though I believe its a decent fit with the literature], using an
expanding-window out-of-sample design.
